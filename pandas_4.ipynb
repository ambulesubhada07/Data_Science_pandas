# **Fifa Worldcup dataset analysis**

```python
import pandas as pd
f = pd.read_csv('https://raw.githubusercontent.com/hakimyameen/Datasets/refs/heads/main/fifa_eda.csv')
f.head(2)
```

```python
f.shape
```

```python
# check unique nationality
f['Nationality'].unique()
```

```python
f['Position'].unique()
```

```python
# count total countries present
f['Nationality'].nunique()
```

```python
# Countrywise count
f['Nationality'].value_counts().head(10)  # top 10 countries w.r.t count
```

```python
sum(f['Nationality'] == 'Portugal')
```

```python
# select the player whose Nationality is Portugal
f[f['Nationality'] == 'Portugal']
```

```python
pd.set_option('display.max_columns',None)
```

```python
f.query('Nationality=="Portugal" and Age>30')
```

```python
f.query('Nationality=="Portugal" and Age>30').shape
```

```python
# count  for Manchester City player
sum(f['Club'] == 'Manchester City')
```

```python
#f[f['Club'] == 'Manchester City']

# using query(str/expression)
f.query('Club=="Manchester City"')
```

```python
# player from Manchester City whose age>35
#f.query('Club=="Manchester City" & Age<20')  # & |
f[(f['Club'] == 'Manchester City') & (f["Age"]<20)]
```

```python
# average of age overall potential wage value
f.loc[:,['Age','Overall','Potential','Value','Wage']].mean()
```

```python
f[:2]
```

```python
# Display average age w.r.t country
f.groupby('Nationality')['Age'].mean()  # sorts data A-Z
```

```python
# averaging with other inputs: give me the max value of Age Value Overall columns w.r.t. each country
f.groupby('Nationality')[['Age','Value','Overall']].max()
```
